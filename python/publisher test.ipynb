```python
import rclpy as rp
from geometry_msgs.msg import Twist

rp.init()
test_node = rp.create_node('pub_test')
```

```python
msg = Twist()
print(msg)
```

```python
pub = test_node.create_publisher(Twist, '/turtle1/cmd_vel', 10)
pub.publish(msg)
```

```python
msg.linear.x= 3.0
msg.angular.z = 0.35
pub.publish(msg)
```

```python
cnt = 0
def timer_callback():
    global cnt
    cnt += 1
    print(cnt)
    pub.publish(msg)
    if cnt > 5:
        raise Exception("Publisher Stop")
```

```python
timer_period = 0.1
timer = test_node.create_timer(timer_period, timer_callback)
rp.spin(test_node)
```
