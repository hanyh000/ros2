# 파이썬코드로 토픽 구독하기

## ROS2토픽  
**토픽**은 발행과 구독으로 되어있는 개념이다.

## ROS Client LIbrary For Python   
**rclpy**는 파이썬을 위한 ROS클라이언트 라이브러리이다.

```python
import rclpy as rp
from turtlesim.msg import Pose
```

## import 하는법  
위 코드를 보면 import rclpy as rp라고 설정한 부분이 있는데 rclpy가 가지고 있는 함수를 실행하려면 rp.init()라고 사용하면 된다.  
from turtlesim.msg import Pose라고 써진 import는 turtlesim.msg하위의 Pose만 별도로 import한 것이다.

```python
rp.init()
test_node = rp.create_node('sub_test')
```

## rclpy의 초기화 및 노드생성  
rp.init()으로 rclpy초기화를 하고 sub_test라는 이름의 노드를 하나 만들기 위해 rp.create_node('sub_test')를 입력하여 노드를 만든다.

```python
def callback(data):
    print("-->") 
    print("/turtle1/pose : ",data)
    print("X : ", data.x)
    print("Y : ", data.y)
    print("Theta : ", data.theta)
```

## subscription에서 실행할 callback함수 정의하기  
**callback**이 받는 데이터는 turtlesim/msg/Pose이다. 구성은 x, y, theta, angular velocity로 구성된다. 이 데이터를 출력하기 위해 callback 함수를 구성했다.

```python
test_node.create_subscription(Pose, '/turtle1/pose', callback, 10)
```

## 토픽 subscriber 만들기  
위에 추가한 코드는 생성한 test_node에서 create_subscription을 test_node.create_subscription(<data_type>, '<topic_name>', <callback>, <QoSHistory>)의 형식으로 호출하면 된다.

```python
rp.spin_once(test_node)
```

## 노드를 실행하여 토픽 받기
위 코드 처럼 지정하면 해당 노드를 구독할 수 있다. 실행을 하면 아래와 같은 결과를 얻을 수 있다.  
--->  
/turtle1/pose :  turtlesim.msg.Pose(x=5.544444561004639, y=5.544444561004639, theta=0.0, linear_velocity=0.0,  
angular_velocity=0.0)  
X :  5.544444561004639  
Y :  5.544444561004639  
Theta :  0.0

```python
rp.spin_once(test_node)
```

```python
rp.spin(test_node)
```

## spin과 spin_once의 차이  
**spin_once**는 토픽을 한번만 받아들이는 명령이고 **spin**은 토픽을 계속 무한히 구독하게 하는 명령이다.
